```python
import pandas as pd
```

```python
df1 = pd.read_excel('cadastro_itens_teste_questao_2.xlsx')
df2 = pd.read_excel('faturamento_teste_questao_2.xlsx')
```

```python
df3 = df2.loc[df2['Código do Produto'].isin(df1['Código do Produto']) == False]
df3
```

```python
prod_nao_cadastrado = df3['Código do Produto']
valor_nao_contabilizado = df3['  Valor líquido']
total_nao_contabilizado = sum(df3['  Valor líquido'])
df4 = pd.DataFrame({'Produtos não cadastrados': prod_nao_cadastrado, 'Valor Faturado não contabilizado': valor_nao_contabilizado, 'Total não contabilizado nos meses de Janeiro/Fevereiro': total_nao_contabilizado })
df4
```
